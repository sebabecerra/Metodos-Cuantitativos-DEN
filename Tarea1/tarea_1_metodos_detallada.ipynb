# Parte 1: Propuesta de escalas de medición

Para cada variable se propone:

- una escala **métrica**, que permite hacer operaciones numéricas más fuertes;
- una escala **no métrica**, que clasifica u ordena respuestas sin asumir distancia exacta entre categorías.

Recordatorio conceptual:

- **Nominal:** clasifica sin orden.
- **Ordinal:** clasifica con orden, pero sin distancias iguales garantizadas.
- **Intervalo:** hay orden y distancias interpretables, pero no cero absoluto.
- **Razón:** hay orden, distancias interpretables y cero absoluto.

## 1.a Intención de recompra

### Escala métrica propuesta

**Tipo de escala:** intervalo.

**Ítem:** "En una escala de 0 a 10, ¿qué tan probable es que vuelva a comprar productos o servicios de esta empresa durante los próximos seis meses?"

**Opciones de respuesta:**

- 0 = Nada probable
- 10 = Muy probable
- El encuestado puede elegir cualquier número entero entre 0 y 10.

**Justificación:** se trata como métrica porque los valores permiten comparar intensidad. La diferencia entre 2 y 4 se interpreta como el mismo salto que entre 7 y 9.

### Escala no métrica propuesta

**Tipo de escala:** ordinal.

**Ítem:** "¿Cuál de las siguientes alternativas describe mejor su intención de volver a comprar en esta empresa?"

**Opciones de respuesta:**

1. Definitivamente no volvería a comprar
2. Probablemente no volvería a comprar
3. No estoy seguro/a
4. Probablemente volvería a comprar
5. Definitivamente volvería a comprar

**Justificación:** las respuestas tienen orden, pero no se puede afirmar que la distancia entre categorías sea exactamente igual.

# Parte 2: Constructos reflectivos y formativos

En medición de constructos, la diferencia clave es la dirección conceptual entre el constructo y sus indicadores.

En una escala **reflectiva**, el constructo latente causa las respuestas observadas. Por eso los indicadores tienden a correlacionarse y pueden ser relativamente intercambiables.

En una escala **formativa**, los indicadores forman o componen el constructo. Por eso no tienen que correlacionarse necesariamente, y eliminar un indicador puede cambiar el significado del constructo.

## 2.a Constructo medido con escala reflectiva: satisfacción del cliente

### Constructo

**Satisfacción del cliente.**

### Definición conceptual

La satisfacción del cliente es la evaluación global que realiza un consumidor respecto de su experiencia con una empresa, producto o servicio, considerando el grado en que el desempeño percibido cumple o supera sus expectativas.

### Escala de medición propuesta

Escala tipo Likert de 1 a 7, donde:

- 1 = Totalmente en desacuerdo
- 7 = Totalmente de acuerdo

Ítems:

1. "Estoy satisfecho/a con mi experiencia general con esta empresa."
2. "La experiencia con esta empresa cumplió mis expectativas."
3. "Mi experiencia con esta empresa fue cercana a lo ideal."
4. "Tomé una buena decisión al elegir esta empresa."

### Justificación de naturaleza reflectiva

La escala es reflectiva porque la satisfacción global del cliente se manifiesta en las respuestas a los ítems. Si una persona está altamente satisfecha, debería tender a estar de acuerdo con todos o la mayoría de los enunciados. Los indicadores son manifestaciones del mismo estado psicológico, por lo que se espera que estén correlacionados. Además, eliminar un ítem no cambia completamente el significado del constructo, aunque puede reducir la precisión de la medición.

## 2.b Constructo medido con escala formativa: imagen de responsabilidad social corporativa

### Constructo

**Imagen percibida de responsabilidad social corporativa, CSR.**

### Definición conceptual

La imagen percibida de responsabilidad social corporativa corresponde a la evaluación que hacen los stakeholders sobre el grado en que una empresa actúa responsablemente frente a distintas dimensiones sociales, laborales, ambientales, éticas y comunitarias.

### Escala de medición propuesta

Escala tipo Likert de 1 a 7, donde:

- 1 = Totalmente en desacuerdo
- 7 = Totalmente de acuerdo

Indicadores formativos:

1. "La empresa se preocupa por reducir su impacto ambiental."
2. "La empresa mantiene buenas prácticas laborales con sus trabajadores."
3. "La empresa contribuye al desarrollo de la comunidad."
4. "La empresa actúa de manera ética y transparente."
5. "La empresa ofrece productos o servicios seguros para sus consumidores."

Una forma simple de operacionalizar el constructo es calcular un índice promedio o ponderado de estas dimensiones.

### Justificación de naturaleza formativa

La escala es formativa porque las distintas dimensiones construyen la imagen global de CSR. Una empresa puede ser fuerte en prácticas ambientales, pero débil en trato laboral o contribución comunitaria. Por eso los indicadores no tienen que moverse todos juntos ni ser intercambiables. Además, eliminar una dimensión, por ejemplo la ambiental o la laboral, cambia el significado del constructo, porque se pierde una parte sustantiva de lo que se entiende por responsabilidad social corporativa.

## 6. Crear la variable de cambio en confianza

Para la pregunta 3.g necesitamos medir cuánto cambió la confianza después de recibir la información.

Definimos:

$$
\text{Cambio en confianza} = \text{Confianza después} - \text{Confianza antes}
$$

Interpretación:

- valor positivo: la confianza aumentó;
- valor cero: la confianza no cambió;
- valor negativo: la confianza disminuyó.

```python
df["Cambio_confianza"] = df["Confianza_después"] - df["Confianza_antes"]

df[["Grupo_label", "Confianza_antes", "Confianza_después", "Cambio_confianza"]].head()
```

## 7. Funciones auxiliares

Estas funciones evitan repetir código.

También ayudan a mantener una interpretación homogénea en todas las pruebas.

```python
def decision_hipotesis(p_value, alpha=ALPHA):
    if p_value < alpha:
        return "Se rechaza H0"
    return "No se rechaza H0"


def interpretar_significancia(p_value, alpha=ALPHA):
    if p_value < alpha:
        return "Existe evidencia estadísticamente significativa al 95% de confianza."
    return "No existe evidencia estadísticamente significativa al 95% de confianza."


def imprimir_resultado(nombre_prueba, estadistico, p_value, gl=None, critico=None):
    print(nombre_prueba)
    print("-" * len(nombre_prueba))
    print(f"Estadístico calculado: {estadistico:.4f}")
    if gl is not None:
        print(f"Grados de libertad: {gl}")
    if critico is not None:
        print(f"Estadístico crítico al 95%: {critico}")
    print(f"p-value: {p_value:.6f}")
    print("Decisión:", decision_hipotesis(p_value))
    print("Interpretación:", interpretar_significancia(p_value))


def resumen_numerico_por_grupo(variable):
    return (
        df.groupby("Grupo_label")[variable]
        .agg(n="count", media="mean", desv_est="std", minimo="min", maximo="max")
        .round(4)
    )
```

## 9. Pregunta 3.a: prueba t de dos muestras para edad

### Objetivo

Comparar la edad promedio entre los dos grupos de tratamiento:

- CSR Externa
- CSR Interna

No se considera el grupo de control.

### Hipótesis

$$
H_0: \mu_{\text{Edad, CSR Externa}} = \mu_{\text{Edad, CSR Interna}}
$$

$$
H_1: \mu_{\text{Edad, CSR Externa}} \neq \mu_{\text{Edad, CSR Interna}}
$$

### Prueba usada

Usamos una prueba t de dos muestras independientes.

En particular, usamos la versión de **Welch**, porque no exige asumir igualdad de varianzas entre los grupos.

```python
edad_externa = df.loc[df["Grupo"] == 1, "Edad"].dropna()
edad_interna = df.loc[df["Grupo"] == 2, "Edad"].dropna()

n1 = len(edad_externa)
n2 = len(edad_interna)

media1 = edad_externa.mean()
media2 = edad_interna.mean()

sd1 = edad_externa.std(ddof=1)
sd2 = edad_interna.std(ddof=1)

print("Descriptivos")
print("------------")
print(f"CSR Externa: n={n1}, media={media1:.4f}, sd={sd1:.4f}")
print(f"CSR Interna: n={n2}, media={media2:.4f}, sd={sd2:.4f}")
```

```python
t_stat, p_value = stats.ttest_ind(
    edad_externa,
    edad_interna,
    equal_var=False
)

var1 = sd1 ** 2
var2 = sd2 ** 2

gl_welch = ((var1 / n1 + var2 / n2) ** 2) / (
    ((var1 / n1) ** 2) / (n1 - 1) + ((var2 / n2) ** 2) / (n2 - 1)
)

t_critico = t.ppf(1 - ALPHA / 2, gl_welch)

t_edad_stat = t_stat
p_edad_value = p_value
gl_edad_welch = gl_welch

imprimir_resultado(
    nombre_prueba="Prueba t de Welch para edad entre tratamientos",
    estadistico=t_stat,
    p_value=p_value,
    gl=round(gl_welch, 4),
    critico=f"±{t_critico:.4f}"
)
```

### Interpretación 3.a

Si el p-value es mayor que 0.05, no hay evidencia suficiente para afirmar que los grupos de tratamiento difieran en edad promedio.

Esto es importante porque sugiere que ambos tratamientos son comparables en edad.

## 10. Pregunta 3.b: prueba t para muestras relacionadas

### Objetivo

Evaluar si la confianza cambia antes vs. después dentro de cada grupo.

Se hace una prueba separada para:

- CSR Externa
- CSR Interna
- Info sin CSR

### Hipótesis para cada grupo

$$
H_0: \mu_{\text{después} - \text{antes}} = 0
$$

$$
H_1: \mu_{\text{después} - \text{antes}} \neq 0
$$

### Prueba usada

Usamos una prueba t pareada porque son las mismas personas medidas dos veces.

```python
resultados_t_pareada = []

for codigo_grupo, nombre_grupo in labels_grupo.items():
    pares = (
        df.loc[df["Grupo"] == codigo_grupo, ["Confianza_antes", "Confianza_después"]]
        .dropna()
    )
    
    diferencias = pares["Confianza_después"] - pares["Confianza_antes"]
    
    n = len(diferencias)
    media_antes = pares["Confianza_antes"].mean()
    media_despues = pares["Confianza_después"].mean()
    media_cambio = diferencias.mean()
    sd_cambio = diferencias.std(ddof=1)
    
    t_stat, p_value = stats.ttest_rel(
        pares["Confianza_después"],
        pares["Confianza_antes"]
    )
    
    gl = n - 1
    t_critico = t.ppf(1 - ALPHA / 2, gl)
    
    print("\n" + "=" * 80)
    print(f"Grupo: {nombre_grupo}")
    print("=" * 80)
    print(f"N pares: {n}")
    print(f"Media antes: {media_antes:.4f}")
    print(f"Media después: {media_despues:.4f}")
    print(f"Media cambio: {media_cambio:.4f}")
    print(f"Desv. est. cambio: {sd_cambio:.4f}")
    
    imprimir_resultado(
        nombre_prueba="Prueba t pareada",
        estadistico=t_stat,
        p_value=p_value,
        gl=gl,
        critico=f"±{t_critico:.4f}"
    )
    
    resultados_t_pareada.append({
        "grupo": nombre_grupo,
        "n": n,
        "media_antes": media_antes,
        "media_despues": media_despues,
        "media_cambio": media_cambio,
        "sd_cambio": sd_cambio,
        "t": t_stat,
        "gl": gl,
        "p_value": p_value,
        "decision": decision_hipotesis(p_value)
    })

tabla_t_pareada = pd.DataFrame(resultados_t_pareada)
tabla_t_pareada.round(4)
```

### Interpretación 3.b

Una prueba significativa indica que la confianza cambió de forma estadísticamente significativa dentro de ese grupo.

La media del cambio permite ver la dirección:

- media positiva: aumentó la confianza;
- media negativa: disminuyó la confianza;
- media cercana a cero: cambió poco.

## 11. Pregunta 3.c: prueba Z de porcentajes para género

### Objetivo

Comparar la proporción de participantes de género femenino entre los dos grupos de tratamiento.

No se considera el grupo de control.

### Hipótesis

$$
H_0: p_{\text{Femenino, CSR Externa}} = p_{\text{Femenino, CSR Interna}}
$$

$$
H_1: p_{\text{Femenino, CSR Externa}} \neq p_{\text{Femenino, CSR Interna}}
$$

### Prueba usada

Usamos una prueba Z para dos proporciones.

```python
tratamientos = df[df["Grupo"].isin([1, 2])].copy()

grupo_externa = tratamientos[tratamientos["Grupo"] == 1]
grupo_interna = tratamientos[tratamientos["Grupo"] == 2]

evento = 2

x1 = (grupo_externa["Genero"] == evento).sum()
x2 = (grupo_interna["Genero"] == evento).sum()

n1 = grupo_externa["Genero"].notna().sum()
n2 = grupo_interna["Genero"].notna().sum()

p1 = x1 / n1
p2 = x2 / n2

print(f"Evento analizado: género femenino = {evento}")
print(f"CSR Externa: {x1} de {n1} = {p1:.4f}")
print(f"CSR Interna: {x2} de {n2} = {p2:.4f}")
```

```python
count = np.array([x1, x2])
nobs = np.array([n1, n2])

z_stat, p_value = proportions_ztest(
    count=count,
    nobs=nobs,
    alternative="two-sided"
)

z_critico = stats.norm.ppf(1 - ALPHA / 2)

z_genero_stat = z_stat
p_genero_value = p_value

imprimir_resultado(
    nombre_prueba="Prueba Z de dos proporciones para género",
    estadistico=z_stat,
    p_value=p_value,
    gl=None,
    critico=f"±{z_critico:.4f}"
)
```

### Interpretación 3.c

Si el resultado no es significativo, no hay evidencia de que la composición por género sea distinta entre los dos tratamientos.

Eso ayuda a evaluar si los grupos son comparables en esta variable.

## 12. Pregunta 3.d: chi-cuadrado para ocupación

### Objetivo

Evaluar si la distribución de ocupación difiere entre los dos grupos de tratamiento.

No se considera el grupo de control.

### Hipótesis

$$
H_0: \text{ocupación y grupo de tratamiento son independientes}
$$

$$
H_1: \text{ocupación y grupo de tratamiento no son independientes}
$$

### Prueba usada

Usamos una prueba de chi-cuadrado de independencia.

Los grados de libertad se calculan como:

$$
gl = (r - 1)(c - 1)
$$

donde `r` es el número de filas y `c` el número de columnas.

```python
tabla_ocupacion = pd.crosstab(
    tratamientos["Grupo_label"],
    tratamientos["Ocupacion_label"]
)

tabla_ocupacion
```

```python
chi2_stat, p_value, gl, expected = stats.chi2_contingency(tabla_ocupacion)
chi2_critico = chi2.ppf(1 - ALPHA, gl)

chi_ocupacion_stat = chi2_stat
p_ocupacion_value = p_value
gl_ocupacion = gl

frecuencias_esperadas_ocupacion = pd.DataFrame(
    expected,
    index=tabla_ocupacion.index,
    columns=tabla_ocupacion.columns
)

print("Frecuencias esperadas:")
display(frecuencias_esperadas_ocupacion.round(4))

imprimir_resultado(
    nombre_prueba="Chi-cuadrado de independencia para ocupación",
    estadistico=chi2_stat,
    p_value=p_value,
    gl=gl,
    critico=f"{chi2_critico:.4f}"
)
```

### Interpretación 3.d

Si el p-value es mayor que 0.05, no hay evidencia suficiente para afirmar que la ocupación esté asociada al grupo de tratamiento.

## 13. Pregunta 3.e: ANOVA para confianza antes y después

### Objetivo

Comparar las medias de confianza entre los tres grupos:

- CSR Externa
- CSR Interna
- Info sin CSR

Se hacen dos ANOVA:

1. Confianza antes de la información.
2. Confianza después de la información.

### Hipótesis general del ANOVA

$$
H_0: \mu_1 = \mu_2 = \mu_3
$$

$$
H_1: \text{al menos una media es diferente}
$$

### Prueba post hoc

Si el ANOVA es significativo, se aplica Tukey HSD para identificar qué pares de grupos difieren.

```python
def anova_una_via(variable):
    grupos = []
    
    print("\n" + "=" * 80)
    print(f"ANOVA para: {variable}")
    print("=" * 80)
    
    for codigo_grupo, nombre_grupo in labels_grupo.items():
        valores = df.loc[df["Grupo"] == codigo_grupo, variable].dropna()
        grupos.append(valores)
        print(
            f"{nombre_grupo}: "
            f"n={len(valores)}, media={valores.mean():.4f}, sd={valores.std(ddof=1):.4f}"
        )
    
    f_stat, p_value = stats.f_oneway(*grupos)
    
    k = len(grupos)
    N = sum(len(g) for g in grupos)
    
    gl_entre = k - 1
    gl_dentro = N - k
    
    f_critico = f.ppf(1 - ALPHA, gl_entre, gl_dentro)
    
    imprimir_resultado(
        nombre_prueba=f"ANOVA de una vía para {variable}",
        estadistico=f_stat,
        p_value=p_value,
        gl=f"{gl_entre}, {gl_dentro}",
        critico=f"{f_critico:.4f}"
    )
    
    if p_value < ALPHA:
        print("\nPost hoc Tukey HSD:")
        temp = df[["Grupo_label", variable]].dropna()
        tukey = pairwise_tukeyhsd(
            endog=temp[variable],
            groups=temp["Grupo_label"],
            alpha=ALPHA
        )
        print(tukey)
    else:
        print("\nNo se aplica Tukey HSD porque el ANOVA no es significativo.")
    
    return {
        "variable": variable,
        "F": f_stat,
        "gl_entre": gl_entre,
        "gl_dentro": gl_dentro,
        "F_critico": f_critico,
        "p_value": p_value,
        "decision": decision_hipotesis(p_value)
    }
```

```python
resultado_anova_antes = anova_una_via("Confianza_antes")
resultado_anova_despues = anova_una_via("Confianza_después")

tabla_anova_confianza = pd.DataFrame([resultado_anova_antes, resultado_anova_despues])
tabla_anova_confianza.round(4)
```

### Interpretación 3.e

El ANOVA de confianza antes permite revisar si los grupos partían desde niveles similares.

El ANOVA de confianza después permite evaluar si, luego de la intervención, existen diferencias entre grupos.

Cuando Tukey HSD aparece, se debe mirar la columna `reject`:

- `True`: ese par de grupos difiere significativamente;
- `False`: ese par de grupos no difiere significativamente.

## 18. Guía para redactar la respuesta final

Para cada prueba, redacta usando esta estructura:

1. **Objetivo:** qué variable o grupos se comparan.
2. **Hipótesis nula:** no hay diferencia o no hay asociación.
3. **Hipótesis alternativa:** sí hay diferencia o sí hay asociación.
4. **Estadístico:** reportar `t`, `Z`, `F` o `chi-cuadrado`.
5. **Grados de libertad:** cuando corresponda.
6. **Valor crítico al 95%:** reportar el estadístico de comparación.
7. **Decisión:** rechazar o no rechazar H0.
8. **Interpretación:** explicar el resultado en palabras del experimento.

Ejemplo de estilo:

> La prueba t pareada para el grupo CSR Externa muestra un cambio significativo en la confianza antes y después de recibir la información. Dado que el p-value es menor a 0.05, se rechaza la hipótesis nula de ausencia de cambio. En términos sustantivos, la información sobre CSR externa aumentó significativamente la confianza promedio en la empresa.
